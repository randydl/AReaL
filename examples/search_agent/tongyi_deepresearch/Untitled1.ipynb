```python
import json
import random
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from tqdm.auto import tqdm
```

```python
def count_turns(messages):
    messages = json.loads(messages)
    messages = [m for m in messages if m['role'] == 'assistant']
    return len(messages)
```

```python
df = pd.read_json('Qwen3-8B.jsonl', lines=True)
df = pd.read_json('epoch2epochstep249globalstep749.jsonl', lines=True)
df['num_turns'] = df['messages'].apply(count_turns)
```
